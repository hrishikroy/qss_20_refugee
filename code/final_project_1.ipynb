```python
# Load Packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from plotnine import * 
from sklearn.linear_model import LinearRegression 
import statsmodels.api as sm
```

```python
asylum_data = pd.read_csv("/Users/hrishikroy/Downloads/Refugee_Asylum_data.csv") 
asylum_data
```

```python
asylum_data = asylum_data[["Year", "Country of origin", "Returned refugees"]] 
asylum_data
```

```python
# Cleaning datasets 

returned_by_year_country = (
    asylum_data
    .groupby(["Year", "Country of origin"], as_index = False)["Returned refugees"]
    .sum()
)

returned_by_year_country 

returned_by_year_country = returned_by_year_country.sort_values(
    ["Country of origin", "Year"]
)

returned_by_year_country
```

```python
violence_data = pd.read_csv("/Users/hrishikroy/Downloads/violence_acled.csv") 
violence_data
```

```python
# Aggregate by year and country
year_country_summary = (
    violence_data
    .groupby(["year", "country"])
    .agg(
        aggregate_casualties=("fatalities", "sum"),
        number_of_events=("event_id_cnty", "nunique")
    )
    .reset_index()
    .sort_values(["year", "country"])
)

year_country_summary
```

```python
# Rename asylum columns to match violence summary columns
returned_by_year_country_clean = returned_by_year_country.rename(columns={
    "Year": "year",
    "Country of origin": "country",
    "Returned refugees": "returned_refugees"
})

# Join violence data with returned refugee data by country and year
combined_year_country = year_country_summary.merge(
    returned_by_year_country_clean,
    on=["country", "year"],
    how="left"
)

combined_year_country
```

```python
unemp_data = pd.read_csv("/Users/hrishikroy/Downloads/unemp.csv") 
unemp_data
```

```python
unemp_long = unemp_data.melt(
    id_vars=["country", "Country Code"],
    var_name="year",
    value_name="unemployment_rate"
)

# Convert year to integer
unemp_long["year"] = unemp_long["year"].astype(int)

# Keep only the columns needed for the join
unemp_long = unemp_long[["country", "year", "unemployment_rate"]]

unemp_long
```

```python
combined_year_country = combined_year_country.merge(
    unemp_long,
    on=["country", "year"],
    how="left"
)

combined_year_country
```

```python
# Regression Model 

# Keep only rows with the variables needed for regression
regression_data = combined_year_country.dropna(
    subset=["aggregate_casualties", "returned_refugees", "unemployment_rate"]
)

# Independent variables: main predictor + control
X = regression_data[["returned_refugees", "unemployment_rate"]]

# Dependent variable
y = regression_data["aggregate_casualties"]

# Add intercept
X = sm.add_constant(X)

# Run OLS
model_control = sm.OLS(y, X).fit()

print(model_control.summary())
```
